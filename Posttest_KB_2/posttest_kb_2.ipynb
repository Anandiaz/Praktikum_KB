### Pertama kita import dulu library yang akan digunakan disini saya menggunakan pandas dan pretty table

### Setelah mengimport library kita read data csv nya, disini saya menggunakan decoder ISO-8859-1 karena ada error jika menggunakan decoder default (UTF-8)

```python
import pandas as pd
from prettytable import PrettyTable

df = pd.read_csv("dataGames.csv",encoding = "ISO-8859-1")
```

```python
df.describe()
```

### syntax untuk menampilkan 10 record pertama

### Menampilkan informasi jumlah record, jumlah attribute, nama attribute, dan jumlah attribute dengan tipe angka

```python
dataset = {
    "Jumlah Record" : len(df),
    "Jumlah Atribut" : len(df.columns),
    "Nama Atribut" : df.columns,
    "Jumlah Atribut Angka" : len(df.select_dtypes('int64')) + len(df.select_dtypes('float64'))
}
ds = PrettyTable()
s=""
for i,j in dataset.items():
    if i == 'Nama Atribut':
        for x in dataset['Nama Atribut']:
            s += x + "\n"
        ds.add_column(i,[s])
    else:
        ds.add_column(i,[j])
print(ds)
```
